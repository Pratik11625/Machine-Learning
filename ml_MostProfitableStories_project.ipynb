# preprocsing pipeline

```python
cat=[]
con=[]
for i in x.columns:
    if x[i].dtypes=='object':
        cat.append(i)
    else:
        con.append(i)
```

```python
cat
```

```python
con
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler,OrdinalEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
```

```python
cat_pip= Pipeline(steps=[('imputer',SimpleImputer(strategy='most_frequent')),('encoder',OrdinalEncoder())])
print('cat_pip',cat_pip)
print('--'*50)
con_pip= Pipeline(steps=[('imputer',SimpleImputer(strategy='median')),('scaler',StandardScaler())])
print('con_pip',con_pip)
print('--'*50)
pre=ColumnTransformer([('cat_pip',cat_pip,cat),('con_pip',con_pip,con)])
pre
```

```python
x_feat=pd.DataFrame(pre.fit_transform(x),columns=pre.get_feature_names_out())
x_feat
```

# split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_feat,y,test_size=10,random_state=19)
```
